```python
def _color_red_or_green(val):
    color = 'red' if val == True else 'green'
    return 'color: %s' % color
```

```python
df_zero_mask = (df == 0)
df_zero_mask.tail(30)
df_zero_mask.tail(25).drop(["ID"],axis=1).style.applymap(_color_red_or_green)
```

```python
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis=1)
# It will give us the boolean value True if the data point contains all the values True
#It means all the valuesin that data point is zero
# we have to remove those datapoints
feature_zero_mask.tail(25)
```

```python
df_clean_1 = df.loc[~feature_zero_mask].copy()
```

```python
df_clean_1.head()
```

```python
df_clean_1['ID'].nunique()
# We can here there is no duplicates 
```

```python
for i in df_clean_1.columns:
    column_name = i
    print(df_clean_1[column_name].value_counts())
```

```python
# we have to remove those not available data in pay_1 column.Its not the correct way of del the datapoints.It results in del of around 4000 points
df_clean_2 = df_clean_1[df_clean_1['PAY_1'] != 'Not available']
```

```python
df_clean_2.info()
# Firstly we check the null values and dtypes of features
```

```python
df_clean_2['PAY_1'] = df_clean_2['PAY_1'].astype('int64')
df_clean_2[['PAY_1']].info()
```

```python
df_clean_2["EDUCATION"].value_counts()
```

```python
# But we have no information regarding education values of 0,5,6.So we make it those values as others as 4.
df_clean_2['EDUCATION'].replace(to_replace=[0, 5, 6], value=4, inplace=True)
df_clean_2['EDUCATION'].value_counts()
```

```python
#  Same Scenario we can see in education column
df_clean_2['MARRIAGE'].replace(to_replace=0, value=3, inplace=True)
df_clean_2['MARRIAGE'].value_counts()
```

```python
#Conversion of numerical data to categorical data for better understanding
df_clean_2['EDUCATION_CAT'] = 'none'
```

```python
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
cat_mapping = {
    1: "graduate school",
    2: "university",
    3: "high school",
    4: "others"
}
```

```python
df_clean_2['EDUCATION_CAT'] = df_clean_2['EDUCATION'].map(cat_mapping)
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
# Coversion of categorical to numerical (One hot encoding)
#It will results in another 4 columns of EDUCATION .
edu_ohe = pd.get_dummies(df_clean_2['EDUCATION_CAT'])
edu_ohe.head(10)
df_with_ohe = pd.concat([df_clean_2, edu_ohe], axis=1)
df_with_ohe[['EDUCATION_CAT', 'graduate school',
             'high school', 'university', 'others']].head(10)

df_with_ohe.to_csv('cleaned_data.csv', index=False)
```
